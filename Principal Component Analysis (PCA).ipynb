```python
Principal Component Analysis (PCA)
```

```python
Performing  Principal Component Analysis with parameters n_component=2,,,,(Reducing the dimentionality of the input)
Thus saving the storage space.
```

```python
import pandas as pd 
import numpy as np

dataset = pd.read_csv("dp2.csv")
feature=['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','X12','X13','X14','X15','X16','X17']
X=dataset[feature]
y=dataset.Y
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
trans=pca.transform(X)
print("Transformed X")
print(trans)
```

```python
Spliting dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(trans,y,test_size=0.25,random_state=0)
print(X_train,X_test,y_train,y_test)
```

```python
Implementing Algorithms
```

```python
#RANDOM FOREST

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train,y_train.ravel())
y_pred= clf.predict(X_train)
from sklearn.metrics import classification_report
report = classification_report(y_train, y_pred)
print(report)
```
